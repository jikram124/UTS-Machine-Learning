1. Berdasarkan Jenis Pembelajaran:
Pembelajaran Terawasi (Supervised Learning): Model dipelajari dari data yang telah dilabeli. Contohnya termasuk klasifikasi dan regresi.
Pembelajaran Tidak Terawasi (Unsupervised Learning): Model belajar dari data tanpa label. Contohnya termasuk pengelompokan (clustering) dan reduksi dimensi.
Pembelajaran Penguatan (Reinforcement Learning): Model belajar melalui interaksi dengan lingkungannya, menerima umpan balik dalam bentuk hadiah atau hukuman. Contohnya termasuk aplikasi dalam game atau robotika.

2. Berdasarkan Tujuan:
Klasifikasi: Memisahkan instance ke dalam kelas-kelas atau kategori tertentu.
Regresi: Memprediksi nilai numerik berdasarkan fitur-fitur yang ada.
Pengelompokan (Clustering): Mengelompokkan instance ke dalam kelompok-kelompok yang homogen.
Reduksi Dimensi (Dimensionality Reduction): Mengurangi jumlah fitur dalam data tanpa kehilangan informasi penting.

3. Berdasarkan Metode Pembelajaran:
Berbasis Instance (Instance-based): Model mempelajari pola dari instance-instance yang telah diberikan.
Berbasis Model (Model-based): Model mempelajari pola dari data untuk menghasilkan suatu fungsi yang bisa mengeneralisasi pola tersebut.

4. Berdasarkan Keterlibatan Manusia:
Online Learning: Model diperbarui secara langsung saat data baru masuk.
Batch Learning: Model diperbarui secara periodik dengan seluruh dataset.
Active Learning: Model meminta intervensi manusia untuk memberikan label pada instance yang dianggap paling bermanfaat.

5. Berdasarkan Struktur Model:
Linier (Linear): Model yang mengasumsikan hubungan linier antara variabel input dan output.
Non-Linier (Non-linear): Model yang mampu menangkap hubungan non-linier antara variabel input dan output.
Pemahaman terhadap taxonomi dalam Machine Learning membantu peneliti dan praktisi untuk memilih pendekatan yang sesuai dengan permasalahan yang dihadapi serta memahami karakteristik dan batasan dari masing-masing pendekatan tersebut.

1) Sempurnakan Data Set Universitas Esa Unggul memiliki detail mahasiswa sebagai berikut :
Ani, Budi, Joni, Jono dan Lono Setiap mahasiswa dating ke kampus mulai dari hari senin sampai
minggu Untuk melakukan kegiatan kuliah Ani menghabiskan biaya 30000 dikampus setiap datang
Budi menghabiskan biaya 35000 dikampus setiap datang Joni menghabiskan biaya 20000
dikampus setiap datang Jono 25000 dan Lono 15000



a) Berapa rata-rata mahasiswa datang pada minggu ini?

b) Kapan biaya tertinggi terjadi?

c) Hari apa biaya lebih dari 110000?


d) Siapa yang paling banyak datang ke kampus?

e) Siapa yang datang pada hari minggu?

f) Berapa biaya tertinggi dan terendah?

g) Berapa frekuensi datang tertinggi dan terendah?

````python
# Data set mahasiswa dan informasi biaya
data_mahasiswa = {
    "Ani": {"Senin": 2, "Sabtu": 5, "Minggu": 2, "Biaya": 30000},
    "Budi": {"Selasa": 3, "Minggu": 2, "Biaya": 35000},
    "Joni": {"Jumat": 2, "Biaya": 20000},
    "Jono": {"Rabu": 4, "Biaya": 25000},
    "Lono": {"Kamis": 1, "Biaya": 15000}
}

# Inisialisasi variabel untuk menyimpan hasil perhitungan
total_biaya_hari = {}
total_biaya = 0
total_mahasiswa = 0
hari_biaya_lebih_dari_110000 = []
jumlah_datang_terbanyak = 0
mahasiswa_datang_terbanyak = ""

# Menghitung total biaya dan biaya per hari
for mahasiswa, detail in data_mahasiswa.items():
    for hari, datang in detail.items():
        if hari != "Biaya":
            total_biaya_hari[hari] = total_biaya_hari.get(hari, 0) + (detail["Biaya"] * datang)
            total_biaya += detail["Biaya"] * datang
            total_mahasiswa += 1

            # Mencari mahasiswa dengan jumlah datang terbanyak
            if datang > jumlah_datang_terbanyak:
                jumlah_datang_terbanyak = datang
                mahasiswa_datang_terbanyak = mahasiswa

# Menentukan hari dengan biaya tertinggi
hari_biaya_tertinggi = max(total_biaya_hari, key=total_biaya_hari.get)

# Menentukan hari dimana biaya lebih dari 110000
for hari, biaya in total_biaya_hari.items():
    if biaya > 110000:
        hari_biaya_lebih_dari_110000.append(hari)

# Menentukan mahasiswa yang datang pada hari Minggu
mahasiswa_hari_minggu = [mahasiswa for mahasiswa, detail in data_mahasiswa.items() if "Minggu" in detail]

# Menentukan biaya tertinggi dan terendah
biaya_tertinggi = max(total_biaya_hari.values())
biaya_terendah = min(total_biaya_hari.values())

# Menentukan frekuensi datang tertinggi dan terendah
frekuensi_datang_tertinggi = max(data_mahasiswa[mahasiswa_datang_terbanyak].values())
frekuensi_datang_terendah = min([min(detail.values()) for detail in data_mahasiswa.values() if isinstance(detail, dict)])

# Menampilkan hasil
print("a) Rata-rata mahasiswa datang per hari:", total_mahasiswa / len(total_biaya_hari))
print("b) Biaya tertinggi terjadi pada hari", hari_biaya_tertinggi, "dengan biaya sebesar", biaya_tertinggi)
print("c) Hari dimana biaya lebih dari 110000:", hari_biaya_lebih_dari_110000)
print("d) Mahasiswa yang paling banyak datang ke kampus adalah", mahasiswa_datang_terbanyak, "dengan", jumlah_datang_terbanyak, "kali datang")
print("e) Mahasiswa yang datang pada hari Minggu:", mahasiswa_hari_minggu)
print("f) Biaya tertinggi adalah", biaya_tertinggi, "dan biaya terendah adalah", biaya_terendah)
print("g) Frekuensi datang tertinggi adalah", frekuensi_datang_tertinggi, "dan frekuensi datang terendah adalah", frekuensi_datang_terendah)
````
